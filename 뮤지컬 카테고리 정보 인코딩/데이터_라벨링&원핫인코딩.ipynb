```python
print(list(cat_dic.keys()))
print(cat_dic.values())
```

```python
df['category'].values
```

```python
cat_lists = []
for i in df['category'].values:
    i = i.replace(' ', '')
    i = i[1:-1]
    i = i.replace("'", "")
    cat_list = i.split(',')
    cat_lists.append(cat_list)
print(cat_lists)
```

```python
blank = []
for cats in cat_lists:
    for cat in cats:
        if cat in list(cat_dic.keys()):
            cat_dic[cat] += 1

        else:
            continue
    
    blank.append(pd.DataFrame(cat_dic, index=[0]).values.reshape(-1).tolist())    
    cat_dic = {'코미디':0, '판타지':0, '미스터리':0, '공포':0, '액션':0, '가족':0, '모험':0, '전쟁':0, '시대극':0, '범죄':0, '로맨스':0, '종교':0, '퀴어':0, 
           '스포츠':0, '정치':0, '법정':0, '재난':0, '성장':0, '소설원작':0, '영화원작':0, '실화기반':0, '우정':0, '청춘':0, '복수':0, '키즈':0, '폭력':0, '선정':0, '드라마':0, '스릴러':0}

print(blank)
```

```python
columns = ['코미디', '판타지', '미스터리', '공포', '액션', '가족', '모험', '전쟁', '시대극', '범죄', '로맨스', '종교', '퀴어', 
           '스포츠', '정치', '법정', '재난', '성장', '소설원작', '영화원작', '실화기반', '우정', '청춘', '복수', '키즈', '폭력', '선정', '드라마', '스릴러']
df_cat_encoding = pd.DataFrame(blank, columns=columns)
df_cat_encoding.head()
```

```python
df_edit = pd.concat([df,df_cat_encoding], axis=1)
df_edit.head()
```
