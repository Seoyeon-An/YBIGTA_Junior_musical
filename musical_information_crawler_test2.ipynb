```python
values=[]
driver = webdriver.Chrome()
for i in url_list:
    driver.get(i)
    time.sleep(1)

    #팝업창 종료
    closebutton = '//*[@id="popup-prdGuide"]/div/div[3]/button'
    try:
        close = driver.find_element(By.XPATH, closebutton)
        close.click()
        time.sleep(1)
    except:
        pass

    #뮤지컬제목
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    title = soup.find('h2', attrs={'class':'prdTitle'})
    title = title.get_text().strip()

    #공연 장소
    place=soup.find('a', attrs={'class':'infoBtn'})
    place=place.get_text().strip()
    place=place[:-5]

    #공연 기간, 공연 시간, 관람연령
    playinfo=soup.findAll('p', attrs={'class':'infoText'})
    duration=playinfo[0].get_text()
    playtime=playinfo[1].get_text()
    playage=playinfo[2].get_text()

    #성별 비율
    gender=soup.findAll('div', attrs={'class':'statGenderValue'})
    male=gender[0].get_text()
    female=gender[1].get_text()
    gender=[float(male[:-1]), float(female[:-1])]

    #연령 비율
    age=soup.findAll('span', attrs={'class':'statGraph'})
    for i in range(len(age)):
        age[i]=str(age[i])
        index1=age[i].find('bar=')
        index2=age[i].find('style')
        age[i]=age[i][index1+5:index2-2]
        age[i]=float(age[i])
    
    #dataframe 생성
    line=[]
    line.append(2023) #연도 바꿔주기 중요!!!!
    line.append(title)
    line.append(place)
    line.append(duration)
    line.append(playtime)
    line.append(playage)
    line.append(gender)
    line.append(age)
    values.append(line)

columns=['year', 'title', 'place', 'duration', 'playtime', 'playage', 'gender_ratio', 'age_ratio']
df = pd.DataFrame(values, columns = columns)
filepath="C:/Users/dongh/OneDrive/바탕 화면/뮤지컬 정보/2023_뮤지컬정보.csv"
df.to_csv(filepath,index=False, encoding="utf-8-sig")
```
